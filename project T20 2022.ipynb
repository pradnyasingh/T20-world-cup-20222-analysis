```python
mprice=df["Runs"].mean()
meprice=df["Runs"].median()
sns.distplot(df["Runs"])
plt.axvline(mprice,color="r")
plt.axvline(meprice,color="g")
plt.grid(True)
plt.show()
```

♦Its a pictorial representation of scores secured in entire T20 match which shows positive skewnwes(mean>median)

♦The average runs scored is dipicted by red line which is 66.220000

♦the middle most value ie meadian of runs is 63.000000

```python
plt.figure(figsize=(20,10))
sns.boxplot(data=df,x="Innings",y="Runs")
plt.tight_layout()
plt.show()
```

♦Runs scored in the 1st inning is comparitively higher than that of 2nd inning

```python
#sub plot
plt.figure(1)
plt.subplot(2,2,1)
df["Innings"].value_counts().plot(figsize=(6,6),kind="bar",color="red")
plt.title("T20 INNINGS")
plt.xlabel("no. of innings played")
plt.ylabel("innings type")

plt.subplot(2,2,2)
df["Result"].value_counts().plot(figsize=(6,6),kind="bar",color="green")
plt.title("T20 Results")
plt.xlabel("Results")
plt.ylabel("classification")

plt.subplot(2,2,3)
df["Team"].value_counts().plot(figsize=(6,6),kind="bar",color="purple")
plt.title("T20 Teams")
plt.xlabel("Names of team")
plt.ylabel("scale")

plt.subplot(2,2,4)
df["Opposition"].value_counts().plot(figsize=(6,6),kind="bar",color="yellow")
plt.title("T20 opposition team")
plt.xlabel("Names of Oppositional team ")
plt.ylabel("scale")

plt.tight_layout()
plt.show()
```
